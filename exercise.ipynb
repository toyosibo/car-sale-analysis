## Data Pre-processing & Feature Engineering

```python
autos.isnull().sum()
```

```python
autos.duplicated().sum()
```

```python
# Fill missing 'mileage' values with mean
autos['mileage'].fillna(autos['mileage'].mean(), inplace=True)
```

```python
# Fill missing categorical features with mode
categorical_columns = ['standard_colour', 'reg_code', 'year_of_registration', 'body_type', 'fuel_type']

for col in categorical_columns:
    mode_value = autos[col].mode()[0]
    autos[col] = autos[col].fillna(mode_value)
    print(col, mode_value)
```

```python
autos.isnull().sum()
```

```python
sns.boxplot(x=autos['price'])
```

```python
autos_price = autos.query("price>9000000")
autos_price['price'].value_counts()
```

```python
autos_price
```

```python
autos = autos[~(autos['price']>9000000)]
autos.info()
```

```python
autos.describe().T
```

```python
current_year = 2023
autos['vehicle_age'] = current_year - autos['year_of_registration']
autos.sample(3)
```

```python
autos.drop(columns="year_of_registration", axis=1, inplace=True)
autos.sample(3)
```

```python
sns.barplot(y='vehicle_age', x='body_type', data=autos)
plt.title('Body Type Across The Age of Vehicles')
plt.ylabel('Vehicle Age')
plt.xlabel('Body Type')
plt.xticks(rotation=45)
plt.show()
```

```python
autos['mileage_quartile'] = pd.qcut(autos['mileage'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
autos.head(2)
```

```python
sns.boxplot(y=autos['mileage_quartile'], x=autos['price'], showfliers=False)
plt.ylabel('Mileage Quartile')
plt.xlabel('Price')
plt.title('Distribution of Price across Mileage Quartile')
```

```python
autos.loc[(autos['mileage_quartile']=='3rd') & (autos['vehicle_condition']=='USED')].count()
```

```python
sns.scatterplot(x='vehicle_age', y='price', data=autos)
plt.xlabel('Vehicle Age')
plt.ylabel('Price')
plt.title('Relationship between Price and Vehicle Age')
```

## Associations and Group Differences Analysis

```python
correlation_matrix = low_mileage_high_efficiency.corr(numeric_only=True)
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.scatterplot(x='mileage', y='price', data=new_autos)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Scatterplot Distribution of New Vehicles across Mileage and Price')
```

```python
average_price_by_fuel = autos_newage.groupby('fuel_type')['price'].mean().round(2)
```

```python
plt.bar(average_price_by_fuel.index, average_price_by_fuel)
plt.xticks(rotation=45)
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.title("Average Price by Fuel Type")
```

```python
average_price_by_colour = autos_newage.groupby('standard_colour')['price'].mean().round(2)
```

```python
plt.bar(average_price_by_colour.index, average_price_by_colour, color='steelblue')
plt.xticks(rotation=90)
plt.xlabel('Standard Colour')
plt.ylabel('Average Price')
plt.title('Average Price by Colour')
```

```python
sns.boxplot(x='crossover_car_and_van', y='price', data=autos_newage, showfliers=False)
plt.xlabel('Crossover Car and Van')
plt.ylabel('Price')
plt.title('Box Plot of Price Across Crossover Car and Van')
plt.show()
```
